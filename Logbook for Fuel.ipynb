```python
def plot_boxplot(df, ft):
    df.boxplot(column=[ft])
    plt.grid(False)
    plt.show()
```

```python
# Define outliers
# IQR = Q3-Q1
# +/- 1.5*IQR

def outliers(df, ft):
    Q1 = df[ft].quantile(0.25)
    Q3 = df[ft].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    ls = df.index[(df[ft] < lower_bound) | (df[ft] > upper_bound)]
    
    return ls
```

```python
index_list = []
for feature in ['total_spen', 'gallons', 'cost_per_gallo',  'litres', 'total_spen']:
    index_list.extend(outliers(df4, feature))
```

```python
 index_list
```

```python
def remove(df4, index_list):
    ls = index_list
    ls = sorted(set(ls))
    df = df4.drop(ls)
    return df
```

```python
df4_cleaned = remove(df4, index_list)
```
